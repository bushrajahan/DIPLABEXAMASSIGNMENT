```python
#First task solving using openCv
from google.colab import drive
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

image_path = '/content/Assignment/DIP_Lab_task_01.png'

# Check if the image file exists
if os.path.exists(image_path):
    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    if img is not None:
        height, width = img.shape
        print(f"Image Dimensions: {width}x{height}")

        hist = cv2.calcHist([img], [0], None, [256], [0, 256])

        min_intensity = np.min(img)
        max_intensity = np.max(img)
        print(f"Intensity Range: {min_intensity} - {max_intensity}")

        plt.figure(figsize=(10, 6))
        plt.title("Frequency Distribution of Intensities")
        plt.xlabel("Pixel Intensity")
        plt.ylabel("Frequency")
        plt.plot(hist)
        plt.xlim([0, 256])
        plt.show()
    else:
        print("Error: Could not read the image. Check the file path.")
else:
    print("Image file does not exist.")
```

```python
#problem-3
import cv2
import matplotlib.pyplot as plt


image = cv2.imread('/content/Assignment/DIP_Lab_task_03.png', cv2.IMREAD_GRAYSCALE)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(image, cmap='gray')

plt.subplot(1, 2, 2)
plt.title("Histogram")
plt.hist(image.ravel(), bins=256, range=(0, 256))
plt.show()


gaussian_blur = cv2.GaussianBlur(image, (5, 5), 1)

median_blur = cv2.medianBlur(image, 5)


plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.title("Original Image")
plt.imshow(image, cmap='gray')

plt.subplot(1, 3, 2)
plt.title("Gaussian Blur")
plt.imshow(gaussian_blur, cmap='gray')

plt.subplot(1, 3, 3)
plt.title("Median Blur")
plt.imshow(median_blur, cmap='gray')
plt.show()
```

```python
#problem-4
import cv2
import numpy as np
import matplotlib.pyplot as plt


image = cv2.imread('/content/Assignment/DIP_Lab_task_04.jpg')
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

brightness_reduction = 50

reduced_brightness = cv2.subtract(image, np.array([brightness_reduction, brightness_reduction, brightness_reduction], dtype=np.uint8))

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(image_rgb)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title("Brightness Reduced Image")
plt.imshow(cv2.cvtColor(reduced_brightness, cv2.COLOR_BGR2RGB))  # Convert for RGB display
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python

#problem-5
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('/content/Assignment/DIP_Lab_task_05.jpg', cv2.IMREAD_GRAYSCALE)


_, binary_thresh = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title("Binary Threshold Segmentation")
plt.imshow(binary_thresh, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```
